```python
# Removing games from current year for incomplete information
df = df[df['Release date'] < "2023"]
```

```python
# Analysing relations between Price and Release dates
```

```python
# Selecting some columns
df4 = df[['AppID', 'Name', 'Release date', 'Price']].copy()
df4.shape
```

```python
# Create "Year", "Month", "Week", "Day" and "yyyy-mm" columns
df4['Year'] = df4['Release date'].dt.year
df4['Month'] = df4['Release date'].dt.month
df4['Day'] = df4['Release date'].apply(lambda x: x.day)
df4["Week"] = df4["Release date"].dt.isocalendar().week
df4['yyyy-mm'] = pd.to_datetime(df4['Release date']).dt.strftime('%Y-%m')

df4.head()
```

```python
#df4 = df4.sort_values("Price", ascending=True)

df4 = df4.sort_values("yyyy-mm")

df4.tail(20)
```

```python
#df4.query['Year'].value_counts()
#df4.value_counts('Year', ascending=True).tail(20)
#df4["Year"].unique(ascending=True)

df4.groupby(['yyyy-mm'])['Price'].value_counts()

df4.tail(20)
```

```python
# Create a visualization
sns.relplot(
    data=df4,
    x="yyyy-mm", y="Price",
)
```

```python
# Number of Games by year
plt.figure(figsize=(19,8))
a =sns.countplot(x=df4['Year'],palette="Paired")
plt.title("Number of Games by year")
plt.xticks(rotation=45)
a.set(frame_on=False)
plt.bar_label(a.containers[0], fmt='%.1f',padding=12.0)
```

```python
# Number of Games by month
plt.figure(figsize=(24,8))
a= sns.countplot(x=df4['Month'], palette="Paired")
plt.title("Number of Games by month")
plt.xticks(rotation=45)
plt.bar_label(a.containers[0],padding=7.0)
a.set(frame_on=False)
```

```python
# Number of Games by day
plt.figure(figsize=(24,6))
a= sns.countplot(x=df4['Day'], palette="Paired")
plt.title("Number of Games by day")
plt.xticks(rotation=45)
plt.bar_label(a.containers[0],padding=7.0)
a.set(frame_on=False)
```

```python
sns.displot(data=df4, x="yyyy-mm", kde=True)
```
